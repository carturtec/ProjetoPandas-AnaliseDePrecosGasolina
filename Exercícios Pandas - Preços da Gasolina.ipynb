### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
#cria uma variável chamada df3["MES"] recebendo DATA FINAL
df3["MES"] = df3["DATA FINAL"].apply(lambda x: x.month)
#cria uma variável chamada df_rio recebendo estado 
df_rio = df3[df3["ESTADO"] == "RIO DE JANEIRO"]
```

```python
df_rio
```

```python
#cria um groupby jogando as colununas e dar um last pra pegar a ultima informação já que tem varias leituras no mesmo mes
df_month_rio = df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA", "MES"]].last()
```

```python
df_month_rio[df_month_rio["MES"] == 12]
```

```python
#Shift desloca o dado mantendo o índice deslocou para baixo mas manteve o índice ANO-MES
df_month_rio[df_month_rio["MES"] == 12].shift(1)
```

```python
#faz filtro apenas nos meses 12. Divido os dados dos meses 12 por ele mesmo, deslogo os dados -1 e multiplico por 100
(df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) * 100
```

```python
#SOMATORIO DAS EVOLUÇÕES PERCENTUAIS Divido os dados dos meses 12 por ele mesmo, deslogo os dados -1 e multiplico por 100
#e soma deslocando apartir do segundo dado rolling(2)
((df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) * 100).rolling(2).sum()
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df3
```

```python
#agrupa e mostra ao maiores e menores valore de revenda por período
df_max = df3.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"]
df_min = df3.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"]
```

```python
df_max
```

```python
df_min
```

```python
#cria um dataframe vazio
df_diff = pd.DataFrame()
```

```python
df_diff
```

```python
#preenchendo dataframe com a diferença entre maior e menor preço
df_diff['abs_diff'] = df_max - df_min
```

```python
df_diff
```

```python
#mostrando a diferença percentual
df_diff['percent_diff'] = (df_max - df_min) / df_min * 100
```

```python
df_diff
```

```python
#preço médio máximo de revenda
#df_max = df3.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"].sort_index()
idx_max = df3.groupby('ANO-MES')["PREÇO MÉDIO REVENDA"].idxmax()
#preço médio máximo de revenda
#df_min = df3.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"].sort_index()
idx_min = df3.groupby('ANO-MES')["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
idx_max
```

```python
idx_min
```

```python
#criando dataframe
df_diff = pd.DataFrame()
```

```python
#preenchendo a tabela
df_diff["abs_diff"] = df_max - df_min
df_diff["perct_diff"] = (df_max - df_min) / df_min * 100
df_diff["max"] = df_max
df_diff["min"] = df_min
```

```python
df_diff
```

```python
#loc localiza o idx_max por estado, preço medio e ano-mes
df3.loc[idx_max, :][["ESTADO", "PREÇO MÉDIO REVENDA", "ANO-MES"]]
```

```python
#encaixa estado max e min usando .values já que tem o tamanho ou os índices iguais
df_diff["ESTADO_MAX"] = df3.loc[idx_max, :]["ESTADO"].values
df_diff["ESTADO_MIN"] = df3.loc[idx_min, :]["ESTADO"].values
```
